**PLotly is used to build interactive charts include feature**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import cufflinks as cf
from plotly.offline import iplot
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import plotly.express as px

cf.go_offline()
# cufflinks.iplot seems to have an internal issue with generating color strings for plotly.
# Using plotly.express for a reliable plot.
fig = px.line(tips, y='total_bill', title='Total Bill Over Time', color_discrete_sequence=['blue'])
fig.show()
```

```python
import plotly.express as px

# cufflinks.iplot seems to have an internal issue with generating color strings for plotly.
# Using plotly.express for a reliable plot.
df_mean_tip = tips.groupby('day')['tip'].mean().reset_index()
fig = px.bar(df_mean_tip, x='day', y='tip', title='Average Tip by Day', color='day')
fig.show()
```

```python
import plotly.express as px

# cufflinks.iplot seems to have an internal issue with generating color strings for plotly.
# Using plotly.express for a reliable plot.
fig = px.scatter(tips, x='total_bill', y='tip', title='Total Bill vs Tip')
fig.show()
```

```python
fig1 = px.box(tips, x='total_bill', y='tip', title='Total Bill vs Tip')
fig1.show()
```

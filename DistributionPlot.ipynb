We use ditribution plot to viualize the distribution of quantitative data.

Let's discuss some plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
#displot/histplot
plt.subplot(1, 2, 1)
sns.histplot(df['total_bill'], kde = True) # kde(Kernel Destiny Estimation) = True use for create curve
plt.subplot(1, 2, 2)
sns.histplot(df['tip'], kde = True)
```

```python
sns.displot(df['total_bill'], bins = 50) # Remember displot and histplot both work same but recommended histplot
```

```python
#jointplot
sns.jointplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter')
```

```python
sns.jointplot(data = df, x = 'total_bill', y = 'tip', kind = 'kde') #to know Kind press Shift + Tab
```

```python
#pairplot
sns.pairplot(df, hue = 'sex') # hue is divide by category
```

```python
sns.pairplot(df, hue = 'size', palette = 'rainbow')
```

```python
#rugplot
sns.rugplot(df['total_bill']) # It tell where some value exist
```

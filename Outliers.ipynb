```python
import numpy as np
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,203])
arr
```

```python
Q1 = np.percentile(arr, 25)
Q1
```

```python
Q3 = np.percentile(arr, 75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
UF = Q3 + (1.5*IQR) # Upper Fence
LF = Q1 - (1.5*IQR) # Lower Fence
UF
```

```python
l = []
for i in arr:
  if i <= UF and i >= LF:
     l.append(i)
arr2 = np.array(l)
arr2
```

```python
print(arr)
print(arr2)
```

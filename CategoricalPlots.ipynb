```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### **Categoical Data Plots**
I'll learn using Seaborn to plot categorical data. Few main plot type of this


*   boxplot

*   violinplot
*   stripplot


*   swarmplot


*   barplot
*   countplot





```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(df['sex'])
```

```python
sns.countplot(x = df['sex'], hue = df['smoker'])
```

```python
sns.barplot(x = df['sex'], y = df['tip']) # by default it shows mean so we use estimator
```

```python
import numpy as np
sns.barplot(x = df['sex'], y = df['tip'], estimator = np.sum)
```

```python
sns.boxplot(x = 'tip', y = 'day', data = df, palette = 'rainbow') #bar data points called outliers
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.stripplot(x='day', y='tip', data = df)
```

```python
sns.swarmplot(x='day', y='tip', data = df)
```

```python
sns.violinplot(x = 'tip', y = 'day', data = df, palette = 'rainbow')
sns.swarmplot(x = 'tip', y = 'day', data = df, palette = 'bright')
```

```
x = np.linspace(0, 5, 11)
x
y = x**2
y
```

```
plt.plot(x, y)
plt.title('Hello')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```
#Subplots in Matplotlib
plt.subplot(2, 2, 1)
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.subplot(2, 2, 2)
plt.plot(x*2, x*2)
plt.subplot(2, 2, 3)
plt.plot(x*2, y**2)
```

```
#Object Oriented Way
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x, y)
axis2 = fig.add_axes([0.2, 0.2, 0.4, 0.3])
axis2.plot(x*6, y*4)
```

```
fig = plt.figure(figsize=(10, 7), dpi=100) #(figsize=(10, 7) Used to set width and height of plot canva || dpi=100 improve quality of canva 100 is default
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x, y)
axis2 = fig.add_axes([0.2, 0.4, 0.3, 0.2])
axis2.plot(x, y)
```

```
#Types of plots
plt.scatter(x,y)
```
